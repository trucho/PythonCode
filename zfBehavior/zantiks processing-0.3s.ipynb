# Bloop

```python
length_trial=30 #30 for 0.3s bin, 70 for 0.1s bin
num_trials = int(len(record_trimmed.index)/length_trial)
trial_types = 4
trial_repeats = int(num_trials/trial_types)
bin_size = 0.3
stim_start = 3 #in seconds (3 for 0.3s bin, 2 for 0.1s bin)
response_time=4 #(2 for 0.3s bin, 5 for 0.1s bin)  distance traveled within (in frames)
response_threshold=8

metadata = pd.DataFrame({'Parameters':['Trial types','# replicates','Frames per replicate','Movement bin size in sec','Stim start time in sec', 'Positive response range in frames', 'Positive response definition in pixels'],
                         'Values':[trial_types,trial_repeats,length_trial,bin_size, stim_start, response_time,response_threshold]  
})
```

# Bloop

```python
#This for loop generates the row labels (index_strength) for the dataframe matrix_sumary 
#and column labels (index_column_rep) for the dataframe matrix_sortbyfish

index_strength = []
index_column_rep = ['Time(sec)']

for x in range(trial_types):
    index_strength += ['s'+str(x+1)]
    for y in range(trial_repeats):
        index_column_rep += ['s'+str(x+1)+' rep'+str(y+1)]

matrix_summary = pd.DataFrame(np.zeros((trial_types,12)), index=index_strength, columns=record_trimmed.columns[5:17])#,'s3','s4'
                                                      
matrix_avg_dist = pd.DataFrame(np.zeros((trial_types,12)), index=index_strength, columns=record_trimmed.columns[5:17])#,'s3','s4'
                                                       
```

```python
from scipy.signal import find_peaks

for col_name in list(record_trimmed.columns[5:17]):


    
    
    matrix_holder = record_trimmed.loc[:,col_name]
    index_title= matrix_holder.index[0:length_trial]
    index_time=[round((x*bin_size)-stim_start,1) for x in list(range(length_trial))]



    matrix_sortbyfish = pd.DataFrame(np.nan, index=index_title, columns=index_column_rep)
    #make columns numbers list(range(num_trials))

    matrix_sortbyfish.iloc[:,0] = pd.Series(index_time)

    for curr_repeat in range(trial_repeats):
        for curr_type in range(trial_types):
            #matrix_sortbyfish
            trial= trial_types*curr_repeat+curr_type
            matrix_col = trial_repeats*curr_type+curr_repeat+1
            start=length_trial*trial
            end=length_trial*(trial+1)
            matrix_sortbyfish.iloc[:,matrix_col] = pd.Series(matrix_holder[start:end].reset_index(level=0, drop=True))
            
            
            #matrix_summary
            activity = 0
            #if int(matrix_sortbyfish.iloc[int(stim_start/bin_size)+1,matrix_col]) > 4 and not(int(matrix_sortbyfish.iloc[int(stim_start/bin_size)-1,matrix_col]) > 0 and int(matrix_sortbyfish.iloc[int(stim_start/bin_size),matrix_col]) > 0):
            if int(matrix_sortbyfish.iloc[int(stim_start/bin_size)-1,matrix_col]) < response_threshold:
                for x in range(response_time):
                    activity = activity + int(matrix_sortbyfish.iloc[x + int(stim_start/bin_size),matrix_col])
            
                 
            if activity>response_threshold:
                matrix_summary.iloc[curr_type][matrix_summary.columns.get_loc(col_name)] += 1
                matrix_avg_dist.iloc[curr_type][matrix_summary.columns.get_loc(col_name)] += activity
            
            peak = []
            #matrix_peaks
            #matrix_sortbyfish.iloc[0:int(stim_start/bin_size),matrix_col]
            #peaks, _ = find_peaks(x, height=0)
            #matrix_peaks_pre.iloc[curr_type][matrix_summary.columns.get_loc(col_name)] = len(peaks)
            
    #matrix_sortbyfish.index.names = ['Time(sec)']
    matrix_sortbyfish.columns.names = ['Stim trial']
    matrix_sortbyfish = matrix_sortbyfish.set_index('Time(sec)')
            
            #save as CSV
    
    #matrix_sortbyfish.to_csv(directory+file_name+"_"+col_name+".csv")

                       
matrix_summary=round(matrix_summary/5,2)
matrix_avg_dist=round(matrix_avg_dist/5,2)

matrix_summary_and_text=pd.concat([matrix_summary,matrix_avg_dist,metadata],sort=False)

matrix_summary_and_text.to_csv(directory+file_name+"_summary.csv")
```
